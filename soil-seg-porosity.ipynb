```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage.filters import threshold_multiotsu
import os
```

```python
otsu_img = skimage.io.imread('../results/otsu.tiff')
otsu_img = otsu_img[:, :, :, 0]
tdmrk_img = skimage.io.imread('../results/tdmrk.tiff')
tdmrk_img = tdmrk_img[:, :, :, 0]
dl_img = skimage.io.imread('../results/dl.tiff')
dl_img = dl_img[:, :, :, 0]
dk_img = skimage.io.imread('../results/dk.tiff')
dk_img = dk_img[:, :, :, 0]

otsu = []
tdmrk = []
dl = []
dk = []
n = len(otsu_img[0].reshape(-1))
for i in range(otsu_img.shape[0]):
    otsu.append(sum(sum(otsu_img[i]==0))/n)
    tdmrk.append(sum(sum(tdmrk_img[i]==0))/n)
    dl.append(sum(sum(dl_img[i]==0))/n)
    dk.append(sum(sum(dk_img[i]==0))/n)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, len(otsu)), otsu, label="Otsu's")
plt.plot(np.arange(0, len(otsu)), tdmrk, label="Kriging")
plt.plot(np.arange(0, len(otsu)), dl, label="Unsupervised")
plt.plot(np.arange(0, len(otsu)), dk, label="DeepKriging")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.set_xticklabels(["Otsu's", "Kriging", "Unsupervised", "DeepKriging"])
plt.boxplot([otsu, tdmrk, dl, dk])
```

```python
print(np.mean(otsu), np.std(otsu))
print(np.mean(tdmrk), np.std(tdmrk))
print(np.mean(dl), np.std(dl))
print(np.mean(dk), np.std(dk))
```
